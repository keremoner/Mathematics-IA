```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import random
```

```python
iterations = 1000000

answers = []
answer = 0


for i in range(iterations):
  state = 1
  while True:
    if random.random()<=0.5:
      state = (state+1)%16
    else:
      state = (state-1)%16
    
    answer+=1
    if state == 9:
      break
  answers.append(answer/(i+1))



```

```python
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 12}

matplotlib.rc('font', **font)
fs = 15
t = range(1, iterations+1)
plt.figure(figsize=(14, 10))



plt.plot(t, answers)
#print(answers)

plt.title(r'Graph of Iteration vs Average Jump', fontsize = fs)
plt.xlabel('Iteration', fontsize = fs)
plt.ylabel('Average Jump (Total Jump/Iteration)', fontsize = fs)

plt.locator_params(axis="x", nbins=25)
plt.locator_params(axis="y", nbins=27)

plt.savefig("Internal.png",dpi=300)

print(answers[iterations-1])
```
